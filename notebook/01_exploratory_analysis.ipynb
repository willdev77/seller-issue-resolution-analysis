```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("../data/raw/BR_SAS_Issue_Resolution_Scorecard_20260314_526970.xlsx", sheet_name='Base')

df.head()
```

```python
df = df.dropna(how='all')
```

```python
df = df[df.iloc[:,0] != 'Issues Summary']
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=[
    "seller_name",
    "topic",
    "sub_topic"
], errors="ignore")
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df["resolution_time_days"] = (
    df["last_modified_date"] - df["created_date"]
).dt.days

df["month"] = df["activity_date"].dt.month
df["year"] = df["activity_date"].dt.year
```

```python
tickets_por_funcionarios = df['alias'].value_counts()

tickets_por_funcionarios.head(10)
```

```python
df['status'].value_counts()
```

```python
ticket_mes = df.groupby(['year', 'month'])['activity_id'].count()

ticket_mes
```

```python
#Grafico

#Ticket por funcionario
tickets_por_funcionarios.plot(kind='bar')

plt.title("Top 10 funcionarios por volume de Tickets" )
plt.xlabel('Funcionario')
plt.xticks(rotation = 45)
plt.ylabel('Quantidade de Tickets')

plt.show()
```

```python
df["resolution_time_days"].describe()
```

```python
df["alias"].value_counts().head(10)
```

```python
df['resolution_time_days'].plot(kind='hist', bins=30)

plt.title("Distribuição do Tempo de Resolução")
plt.xlabel('Dias')
plt.ylabel('Quantidade de Tickets')

plt.show()
```

```python
df.groupby('alias')['resolution_time_days'].mean().sort_values()
```

```python
tempor_por_funcionario = df.groupby('alias')['resolution_time_days'].mean()

tempor_por_funcionario.sort_values(ascending=False).plot(kind='bar')

plt.title("Tempo Medio de Resolucao por Funcionario")
plt.xlabel('Funcionario')
plt.xticks(rotation=45)
plt.ylabel('Dias')
```

```python
df.groupby('month_name')['activity_id'].count().sort_values()
```

```python
ticket_mes.plot(kind='line')

plt.title("Tickets ao longo do tempo")
plt.xlabel('Tempo')
plt.xticks(rotation=45)
plt.ylabel('Quantidade')
plt.show()
```

```python
perfomance = df.groupby('alias').agg({
    "activity_id": "count",
    "resolution_time_days": "mean"
})

perfomance = perfomance.sort_values('activity_id', ascending=False)

perfomance
```
